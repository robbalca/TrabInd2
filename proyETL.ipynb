```python
import pandas as pd 
import numpy as np
```

```python
d= pd.read_csv('AccidentesAviones.csv', index_col=0)
```

```python
d.head(4)
```

```python
# se reemplazan los signos ? por nulos

d.replace("?",np.nan, inplace=True)
```

```python
# Cambiar fecha por el formato YY-mm-dd

d[['mes','dia','año']] = d['fecha'].str.split(' ', expand=True)
```

```python
d['dia'] = d.dia.str.replace(',', '')
```

```python
d['mes'] = pd.to_datetime(d['mes'], format= '%B').dt.month
```

```python
d['fecha'] = d['año'].astype(str)+'-'+d['mes'].astype(str)+'-'+d['dia'].astype(str)
```

```python
d.head(4)
```

```python
# transformar la columna Hora_accidente en horas y minutos

d[d['HORA declarada'].str.len()>4]
```

```python
d['HORA declarada'] = d['HORA declarada'].str.replace(':', '')
```

```python
d['HORA declarada'] = d['HORA declarada'].str.replace('c', '')
```

```python
d['HORA declarada'] = d['HORA declarada'].str.replace('Z', '')
```

```python
d['HORA declarada'] = d['HORA declarada'].str.replace(';', '')
```

```python
d['HORA declarada']= d['HORA declarada'].str.strip()
```

```python
# Se reemplaza HORA declarada por hora

d['hora']=pd.to_datetime(d['HORA declarada'], format='%H%M').dt.time
```

```python
# Se elimina HORA declarada

d.drop(columns='HORA declarada', inplace=True)
```

```python
d.head(3)
```

```python
# Eliminar columnas que no voy a utilizar

d = d.drop(columns = ['registration','cn_ln','route'])
```

```python
# Cambiar el nombre de las columnas

d = d.rename(columns = {'hora':'Hora_accidente','OperadOR':'Operador','flight_no':'Numero_vuelo','ac_type':'Tipo_nave','all_aboard':'Total_abordo'})
d = d.rename(columns = {'PASAJEROS A BORDO':'Pasajeros_abordo','crew_aboard':'Tripulacion_abordo','cantidad de fallecidos':'Total_fallecidos','passenger_fatalities':'Pasajeros_fallecidos'})
d = d.rename(columns = {'crew_fatalities':'Tripulacion_fallecida'})
```
